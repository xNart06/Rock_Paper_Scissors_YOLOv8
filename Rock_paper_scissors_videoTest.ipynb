```python
from ultralytics import YOLO
import cv2

# Model dosyasını yükle
model_path = "best.pt"  # Model dosyanızı Colab'e yükleyin
model = YOLO(model_path)

# Video dosyasını yükle
input_video_path = "test.mp4"  # İşlenecek video dosyanızın yolu
output_video_path = "output.mp4"  # İşlenmiş videonun kaydedileceği yol

# Videoyu aç
cap = cv2.VideoCapture(input_video_path)

if not cap.isOpened():
    print("Video açılamadı!")
    exit()

# Video özelliklerini al
fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_size = (frame_width, frame_height)

# Video yazıcı başlat
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # MP4 formatı
out = cv2.VideoWriter(output_video_path, fourcc, fps, frame_size)

# Video üzerinde işlem yapma döngüsü
while True:
    ret, frame = cap.read()
    if not ret:
        print("Video işleme tamamlandı.")
        break

    # Modeli çalıştır ve tahmin al
    results = model.predict(source=frame, conf=0.5, show=False)  # Tahmin eşiğini ayarlayın
    annotated_frame = results[0].plot()

    # İşlenmiş kareyi çıktı videosuna yaz
    out.write(annotated_frame)

# Kaynakları serbest bırak
cap.release()
out.release()

print(f"İşlenmiş video kaydedildi: {output_video_path}")

```
